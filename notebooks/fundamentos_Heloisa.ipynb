```python
df_discursos = pd.read_json(f'{caminho_completo}/bd/BD_BRASIL_GOVFEDERAL_BIBLIOTECA_PRESIDENCIA-bolsonaro-discursos_05.json', orient='columns')
df_discursos = pd.json_normalize(df_discursos['_default'])
# df_discursos
```

#Pré análise

```python
# Indicação da quantidade de linhas e colunas do dataframe
df_discursos.shape
```

```python
# Indicação das colunas do dataframe
df_discursos.columns
```

```python
# Delimitação das colunas a serem utilizadas
df_discursos = df_discursos[['data','titulo','paragrafos']]
df_discursos
```

```python
# filtragem de coluna a partir de uma string
discursos_sem_data = df_discursos[df_discursos["data"]=="08/12/2020"]
discursos_sem_data
```

```python
# https://archive.is/IeQ7I
# conversão da coluna data de string para datetime
df_discursos['data'] = pd.to_datetime(df_discursos['data'], format = '%d/%m/%Y', errors = "coerce")
df_discursos['ano'] = pd.to_datetime(df_discursos['data']).dt.year
discursos_por_ano = df_discursos['ano'].value_counts().sort_index()
discursos_por_ano = discursos_por_ano.to_frame()
discursos_por_ano
```

# Gráficos


```python
discursos_por_ano = px.bar(discursos_por_ano, orientation="h")
discursos_por_ano
```
